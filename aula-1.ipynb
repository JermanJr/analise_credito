````python
cadastros = pd.read_csv('https://github.com/alura-tech/alura-tech-pos-data-science-credit-scoring-streamlit/blob/main/dados/clientes_cadastrados.csv?raw=True')

````

## Caso a regra de negócio entenda que os **últimos** cadastros da base é que é válido desse modo


`O parâmetro keep é usado para determinar qual ocorrência de valores duplicados deve ser marcada como False`

````python
filtro_duplicados = cadastros.ID_Cliente.duplicated(keep='last')
indice_c1_duplicados = cadastros.loc[filtro_duplicados].index
cadastros_c1 = cadastros.drop(index=indice_c1_duplicados)
````

````python
cadastros_c1.shape
````

## Caso a regra de negócio entenda que os **TODOS DUPLICADOS NAO DEVEM APARECER**  e por essa que vou seguir

````python
filtro_duplicados_full = cadastros.ID_Cliente.duplicated(keep=False)
indice_c2_duplicados = cadastros.loc[filtro_duplicados_full].index
df_cadastros_c1 = cadastros.drop(index=indice_c2_duplicados)
df_cadastros_c1.shape
````

## Base duplicados
Para analise de negócio, se devem ou não passar por uma reanalise.

````python
base_duplicados = cadastros[cadastros.ID_Cliente.duplicated(keep=False)].sort_values(by='ID_Cliente')
ids_duplicados = list(set(base_duplicados.ID_Cliente))
````
